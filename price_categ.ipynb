```python
df.info()
```

```python
df['price_string']= df['price_string'].replace('[\$,]', '', regex=True).astype(str) # removing all the currency symbol  

df.head(10)
```

```python
df['price_string']=df['price_string'].apply(lambda x: str(x).replace(x,'$'+str(x))) # when we have completely removed all the dollar sign then again assigning the dollar to the price column

df.head(10)
```

```python
df['price_string']=df['price_string'].apply(lambda x: str(x).replace('$nan',str(np.NaN))) # after assiging replacing all the imporper values with null

df.head(10)
```

```python
df['Currency']=df['price_string'].str[:1] # partitioning the currency and price amount.

df.head(10)
```

```python
df['Currency']=df['Currency'].apply(lambda x: str(x).replace('n',str(np.NaN))) # replacing all the improper values with the null value

df.head(10)
```

```python
df['price_string']= df['price_string'].replace('[\$,]', '', regex=True).astype(str) # As to calculate the average removing all the currency symbol from the price column

df.head(10)
```

```python
df['price_string']=df['price_string'].astype(float) # converting price to float
```

```python
price_avg=df.groupby('category').apply(lambda x : pd.DataFrame(x['price_string']).mean(skipna=True)) # calculating average of prices based on the category

price_avg
```

**Task-5**

```python
price_avg.reset_index(inplace=True) # resetting the index to plot the above dataframe
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.subplots()
ax=sns.barplot(data=price_avg,x='category',y='price_string',hue='category',dodge=False)
sns.move_legend(ax,loc='upper right')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90, horizontalalignment = 'center')
plt.show()
```
